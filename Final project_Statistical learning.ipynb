## Feature Engineering

Categorical variables cannot be directly used by many machine learning algorithms.

Therefore, one-hot encoding is applied to transform categorical variables into numerical representations.

```python
# Before modeling the data, we need to convert categorical feature to numerical value. I am here using one-hot-encoding method 
# Convert all boolean columns to integers (True → 1, False → 0)
data = pd.get_dummies(data, columns=["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"], drop_first=True)
# Convert only boolean columns to integers (True/False → 1/0)
bool_cols = data.select_dtypes(include='bool').columns
data[bool_cols] = data[bool_cols].astype(int)

data.head()
```

```python
#determining the predictor feataurs and target feature 
X = data.drop(columns= ["HeartDisease"], axis = 1)
y = data[["HeartDisease"]]
```

## Model Development

Multiple classification models are implemented to compare predictive performance.

Models tested:

- Logistic Regression (baseline statistical model)
- Decision Tree
- Random Forest
- Support Vector Machine
- K-Nearest Neighbors

The goal is to evaluate how different modeling approaches perform on the same dataset.

## Model Evaluation

Model performance is evaluated using standard classification metrics such as:

- Accuracy
- Confusion Matrix
- Precision
- Recall
- ROC-AUC (if calculated)

These metrics allow comparison of model performance and identification of the most suitable predictive model.

```python
# Implementing logistic model becuase the independent variable is binary and logistic model will be the first logical option.
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler

# Scale the features (important for logistic regression because this model is highly senstive to scale of data)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Flatten y to 1D array
y = y.values.ravel()

# Define number of folds
kf = KFold(n_splits=12, shuffle=True, random_state=42)

# Lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []

# Perform k-fold cross-validation
for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))

# Display results
print("🔹 Logistic Regression Performance (12-Fold CV)")
print(f"Mean Accuracy:  {np.mean(accuracy_scores):.4f}")
print(f"Mean Precision: {np.mean(precision_scores):.4f}")
print(f"Mean Recall:    {np.mean(recall_scores):.4f}")
```

```python
# Implementing decision tree model
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Scale the features (important for logistic regression)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Define number of folds
kf = KFold(n_splits=12, shuffle=True, random_state=42)

# Lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []

# Perform k-fold cross-validation
for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = DecisionTreeClassifier(max_depth=4)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))

# Display results
print("🔹 Decision tree Performance (12-Fold CV)")
print(f"Mean Accuracy:  {np.mean(accuracy_scores):.4f}")
print(f"Mean Precision: {np.mean(precision_scores):.4f}")
print(f"Mean Recall:    {np.mean(recall_scores):.4f}")
```

```python
# Implemendting Random Forest model for classification. 
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Define number of folds
kf = KFold(n_splits=12, shuffle=True, random_state=42)

# Lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []

# Perform k-fold cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]  # FIXED: use slicing since y is ndarray

    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))

# Display results
print("🔹 Random Forest Performance (12-Fold CV)")
print(f"Mean Accuracy:  {np.mean(accuracy_scores):.4f}")
print(f"Mean Precision: {np.mean(precision_scores):.4f}")
print(f"Mean Recall:    {np.mean(recall_scores):.4f}")
```

```python
# Implementing support vector machine model for classification
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler

# Scale X (important for SVM!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define number of folds
kf = KFold(n_splits=12, shuffle=True, random_state=42)

# Lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []

# Perform k-fold cross-validation
for train_index, test_index in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = SVC(random_state=42)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))

# Display results
print("🔹 SVM Performance (12-Fold CV)")
print(f"Mean Accuracy:  {np.mean(accuracy_scores):.4f}")
print(f"Mean Precision: {np.mean(precision_scores):.4f}")
print(f"Mean Recall:    {np.mean(recall_scores):.4f}")
```

```python
# Implementing K-nearest neibour model for classification. 
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score

# Define number of folds
kf = KFold(n_splits=12, shuffle=True, random_state=42)

# Lists to store metrics
accuracy_scores = []
precision_scores = []
recall_scores = []

# Perform k-fold cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]  # y is assumed to be a numpy array

    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))

# Display results
print("🔹 k-NN Classifier Performance (12-Fold CV)")
print(f"Mean Accuracy:  {np.mean(accuracy_scores):.4f}")
print(f"Mean Precision: {np.mean(precision_scores):.4f}")
print(f"Mean Recall:    {np.mean(recall_scores):.4f}")
```
